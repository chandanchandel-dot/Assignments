```python
a = int(input('Enter any number:'))
if a > 1 :
   for i in range(2, a):
       if (a % i) == 0:
           print(f"{a} is not Prime Number")
           break
   else: 
       print(f"{a} is a Prime Number")
else:
    print(f"{a} is not Prime Number")
```

```python
import random
a = random.randint(1,10)
b = random.randint(1,10)
num_mul = a*b
print("Random value of a=",a)
print("Random value of a=",b)
user_answer =int(input("Enter the value of a*b :"))
if  user_answer == num_mul:
    print(f"The user's answer:{user_answer} is correct")
else:
    print(f"The user's answer :{user_answer} is incorrect and correct answer is:{num_mul}.")
```
