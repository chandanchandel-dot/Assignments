```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(model, x_test_set, y_test, model_name):
    
    predictions = model.predict(x_test_set)
    
    # Calculate Metrics
    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    r2 = r2_score(y_test, predictions)
    
    # Calculate Adjusted R-squared
    n = len(y_test) # Number of observations
    p = x_test_set.shape[1] # Number of predictors
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    
    print(f"--- {model_name} Performance ---")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R-squared: {r2:.4f}")
    print(f"Adjusted R-squared: {adj_r2:.4f}\n")

# Evaluating all 3 models built in Task 3
evaluate_model(model1, x_test, y_test, "Model 1 (All Features)")
evaluate_model(model2, x_test[features2], y_test, "Model 2 (Some feature Only)")
evaluate_model(model3, x_test[features3], y_test, "Model 3 (Physical/Utility)")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Function to calculate metrics
def get_metrics(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    return {
        "Model": model_name,
        "MAE": round(mae, 2),
        "RMSE": round(rmse, 2),
        "R2 Score": round(r2, 4)
    }

# Generate predictions for the regularized models
ridge_preds = ridge_model.predict(x_test)
lasso_preds = lasso_model.predict(x_test)

# Collect and display results
results = [
    get_metrics(y_test, ridge_preds, "Ridge Regression"),
    get_metrics(y_test, lasso_preds, "Lasso Regression")
]

perf_df = pd.DataFrame(results)
print(perf_df)
```
