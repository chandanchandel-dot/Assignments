```python
df=df.apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
print(df)
```

```python
df.isnull().sum()
```

```python
df = df.apply(pd.to_numeric, errors='coerce') #Convert all data to numeric
```

```python
print(df)
```

```python
for col in df.select_dtypes(include=[np.number]).columns:
    upper_limit = df[col].quantile(0.99)
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
```

```python
IQR_df = df.describe().T
IQR_df['IQR'] = IQR_df['75%']-IQR_df['25%']
print(IQR_df[['mean','50%', 'std', 'IQR']])
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['LB'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Fetal Heart Rate Baseline (LB)')

sns.histplot(df['ASTV'], kde=True, ax=axes[1], color='orange')
axes[1].set_title('Distribution of % Time with Abnormal STV')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

corr_matrix = df.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Cardiographic Features')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='UC', y='DL', size='ASTV', hue='ASTV', palette='viridis')
plt.title('Impact of Uterine Contractions on Late Decelerations')
plt.xlabel('Uterine Contractions (UC)')
plt.ylabel('Late Decelerations (DL)')
plt.show()
```

```python
main_feature_heartrate = ['LB', 'AC', 'UC', 'ASTV', 'DP']

sns.pairplot(df[main_feature_heartrate], diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Pair Plot of Cardiographic Indicators', y=1.0)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['LB'].rolling(window=30).mean(), label='Mean of Baseline FHR', color='green')
plt.xlabel('Recording Progress (Samples)')
plt.ylabel('Heart Rate')
plt.legend()
plt.show()
```
