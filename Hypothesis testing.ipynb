```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df  = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 
                 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
confid_level = 0.99
sam_mean = np.mean(df)
sam_std = np.std(df)
print(f"Sample Mean: {sam_mean:.2f}")
print(f"Sample Std Dev: {sam_std:.2f}")
```

```python
n = len(df)
print(n)
data = n - 1
import math
ma_criti = stats.t.ppf((1 + 0.99) / 2, data)
margin_of_error = ma_criti * (sam_std / math.sqrt(data))
print(margin_of_error)

ci_t_lower = sam_mean - margin_of_error
ci_t_upper = sam_mean + margin_of_error

print("Sample Std Dev (t-distribution)")
print(f"t-critical value: {ma_criti:.4f}")
print(f"99% Confidence Interval: [{ci_t_lower:.4f}, {ci_t_upper:.4f}]\n")
```

```python
pop_std = 0.2
z_criti = stats.norm.ppf((1 + 0.99) / 2)
print(f"Z Critical: {z_criti: .4f}")
margin_of_error_z = z_criti * (pop_std / math.sqrt(n))
print(f"Margin of error: {margin_of_error_z:.3f}")
lower_range = sam_mean - margin_of_error_z
upper_range = sam_mean + margin_of_error_z
print(f"99% Confidence Interval: [{lower_range:.4f}, {upper_range:.4f}]\n")

```
