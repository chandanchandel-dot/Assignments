```python
age = int(input(f"Enter your age = "))
if age >= 18:
    print('*'*40)
    print(f"You are eligible to cast vote")
    print('*'*40)
else:
    print('*'*40)
    print(f"You are not eligible to cast vote")
    print('*'*40)
```

```python
marks = float(input(f"Enter Your marks : "))
def grade(marks):
    if marks >= 90:
        print(f"Your grade is A")
    elif marks >= 75 and marks <= 89:
        print(f"Your grade is B")
    elif marks >= 50 and marks <= 74:
        print(f"Your grade is C")
    else:
        print(f"Your are Fail")
grade(marks)
```

```python
n = float(input(f"Enter any Number : "))
if n > 0:
    if n % 2 == 0:
        print(f"The number is positive and even")
    else:
        print(f"The number is positive and odd")
elif n == 0:
    print(f"The number is zero")
else:
    print(f"The number is negative")
```
